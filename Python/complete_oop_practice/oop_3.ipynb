### Inheritance

#### What is Inheritance?
- Basic logic of inheritrance (Parent --> child)
- Class_1 --> common methods used by all other classes
- Class_2 --> Unique methods which only class 2 has
- Class_3 --> Unique methods which only class 3 has
- Now the class 2 and class 3 can inherit the common methods from common class "class_1"

#### Why Inheritance
- Code reusability
- Helps in avoiding "DRY" --> "Don't Repeat Yourself" while writing the program

- Triangle means "Inheritence"
- which ever side is the triangle's head that is parent class

#### What can we access as a child class from parent class, what gets inherited

- Constructor
- Non Private Attributes
- Non Private Methods

```python
class Parent:

    def __init__(self):
        self.name = 'ramesh'
        self.gender = 'male'

    def login(self):
        print("login")
    
class Child(Parent):             # Syntax to inherit from Parent class

    def enroll(self):
        print('enrolled into some course')
```

```python
p = Parent()
c = Child()

print(c.name)
print(c.gender)
c.login()
c.enroll()
```

- See the child class doesn't have constructor, it acessed all non_private variables, also accessed the non_private methods

- If there will be a constructor in child class, then it will never call parent's constructor, in that case "child_class.parent's_attribute" will throw error, see below

```python
class A:
    def __init__(self):
        self.var1=100

    def display1(self,var1):
        print("class A :", self.var1)
class B(A):
  
    def display2(self,var1):
        print("class B :", self.var1)

obj=B()
obj.display1(200)
```

### Method Overriding

```python
class Parent:

    def __init__(self):
        print("inside PARENTS's constructor")

    def func(self):
        print("insied PARENT's method")

class Son(Parent):

    def func(self):
        print("insied SON's method")

s = Son()
s.func()
```

- See, because son did not have any constructor, it went to parent class's constructor, but becasue both parent and son had same function, the method of son was executed. This method is known as "Method overriding"

- If we remember even same thing happend with constructor also, when both had constructor, the constructor of child was executed

### Super Keyword

- USed for calling parent's members

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")
        # syntax to call parent's buy method
        super().buy()

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

- See becasue we used " super() " keyword, it called the parent's buy method

- Main usage of super() keyword for constructor

```python
class Phone:
    def __init__(self, price,brand,camera):
        print("second here")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        print("first here")
        super().__init__(price,brand,camera)
        self.os = os
        self.ram = ram
        print("at the end here")

s = SmartPhone(price=20000,brand="Samsung",camera=12,os="Android",ram=2)

print(s.os)
print(s.brand)
```

- We can clearly understand the flow of our program by seeing the output
- Because we had super() keyword, the parent's constructor was also called

#### Inheritance summary

- A class can inherit from another class.

- Inheritance improves code reuse

- Constructor, attributes, methods get inherited to the child class (public)

- The parent has no access to the child class

- Private properties of parent are not accessible directly in child class

- Child class can override the attributes or methods. This is called method overriding

- super() is an inbuilt function which is used to invoke the parent class methods and constructor

```python
# can have multiple except blocks with a generic block in the end for an unexpected error, use this when you want to show 
#  specific error message to user 
try:
    m=2
    print(m)
    print(5/3)
    with open('sample.txt','r') as f:
        f.read()
    l = [1,2,3,4]
    print(l[1])
except NameError:
    print('name error')
except ZeroDivisionError:
    print("can't divide by zero")
except FileNotFoundError:
    print('file nahi mili')
except Exception as e:
    print(e)
```

```python
# Can have a generic except block which catches all errors
try:
    m=2
    print(m)
    print(5/3)
    with open('sample.txt','r') as f:
        f.read()
    l = [1,2,3,4]
    print(l[1])
except Exception as e:
    print(e)
```

```python
# We can include else in such a way that if nothing is going wrong in the try, else will execute
# if something goes wrong except will catch and else will not execute
try:
    f = open('sample.txt','r')
except Exception as e:
    print(e)
else:
    print(f.read())
```

```python
# a laymen example on how to use raise 

class Bank:

    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amount):
        try :
            if amount <0:
                raise Exception("Amount can't be nagative")
            elif self.balance < amount:
                raise Exception(f'Bhai {amount} jithna tho balance bhi nai hai')
        except Exception as e:
            print(e)
        else:
            self.balance = self.balance - amount
            print(f"{amount} is withdrawn, remaining balance is {self.balance}")
```

```python
obj = Bank(10000)
```

```python
obj.withdraw(100000)
```

```python
obj.withdraw(-4000)
```

### We can make our custom exceptions as well

```python
# Same example using cusotmexception

class CustomException(Exception):   # always inherit the Exception class
     def __init__(self,message):
         print(message)


class Bank:

    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amount):
        try :
            if amount <0:
                raise CustomException("Amount can't be nagative")
            elif self.balance < amount:
                raise CustomException(f'Bro {amount} is too high compare to balance')
        except CustomException as e:
            pass # no need to write a print statement "Basically we can customize the message like line numeber file name an etc"
        else:
            self.balance = self.balance - amount
            print(f"{amount} is withdrawn, remaining balance is {self.balance}")
```

```python
obj = Bank(10000)
```

```python
obj.withdraw(100000)
```

```python
obj.withdraw(-4000)
```

```python
class Securityerror(Exception):

  def __init__(self,message):
    print(message)

  def logout(self):
    print('Logout from all devices')


class XYZcompany:

  def __init__(self,name,email,password,device):
    self.name = name
    self.email = email
    self.password = password
    self.device = device

  def login(self,email,password,device):
    if device != self.device:
      raise Securityerror('!!! Something is not correct')
    if email == self.email and password == self.password:
      print('welcome')
    else:
      print('login error')



obj = XYZcompany('Ramesh','ramesh@123@xyz.com','9876','android')

try:
  obj.login('ramesh@123@xyz.com','9876','windows')
except Securityerror as e:
  e.logout()
else:
  print(obj.name)
finally:
  print('database connection closed')

```
